```python
def get_gemini_response(input_prompt,job_titles):
    model=genai.GenerativeModel('gemini-pro')
    print(model)
    response=model.generate_content([input_prompt,job_titles])
    return response.text
```

```python
input_prompt = f'''
Categorize these job titles of same type in one category 

The stucture of the response should be like this Junior Data Analyst=Data Analyst|DATA ANALYST=Data Analyst|Big data=Pyspark| ALM Users - Data Mining=Data Mining ........
the reponse should be in same order as below job-titles, there would be 60 or 25 job-titles so the response whould also has only 60 or 25 job-titles  separate job-title using |, please recheck if the response is in same order like job-title without skipping any thing
job-titles: 
'''
```

```python
print(input_prompt)
```

```python
import time
```

```python
job_column = []
for i in range(0, len(index_list)-1):
    job_titles = df.iloc[index_list[i]:index_list[i+1], 1]
    job_titles = (job_titles).tolist()
    job_titles = "["+' |'.join(job_titles)+"]"
    validate = True
    while validate:
        result = get_gemini_response(input_prompt,job_titles)
        data_list = result.split('|')
        if (len(data_list) == 60 or len(data_list) == 25):
            validate = False
        else:
            sleep_duration = 3
            print(f"Waiting for {sleep_duration} seconds...")
            time.sleep(sleep_duration)
    job_column.append(data_list)
```
